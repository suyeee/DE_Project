```python
# 워드 클라우드 테스트
from wordcloud import WordCloud
from konlpy.tag import Okt
from konlpy.tag import Kkma
from collections import Counter
```

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina' #화질 높이기

text = open('hashtag.txt').read() 

okt = Okt()

# Okt함수를 통해 읽어들인 내용의 형태소를 분석
sentences_tag = []
sentences_tag = okt.pos(text) 

hashtag_list = []

# 워드 클라우드에서 제외할 키워드
unselect_list = ['#cgv', '#daily','#dp','#fff','#followforfollowback','#jtbcdrama','#kdrama', '#myliberationnotes','#가족스타그램','#구자경','#극장데이트','#나의해방일지','#데이트','#데일리','#드라마추천','#디피', '#럽스타그램'
,'#맞팔','#먹방','#먹스타그램','#멜로가체질','#범도2무대인사','#부부','#부부스타그램','#부부데이트','#부산','#비오는날','#선팔','#선팔하면맞팔','#셀카','#셀피','#소통','#소통스타그램','#손석구배우','#심야데이트' ,'#염미정'
,'#영화데이트','#일상스타그램','#일상','#일상기록','#일상그램','#일요일','#오운완','#육아맘','#육아스타그램','#좋반','#주말데이트','#주말일상','#집밥','#천만가즈아','#친정찬스','#현충일']


for word, tag in sentences_tag:
    word = word.replace("[#", "#")

    if tag in ['Hashtag'] and (word not in unselect_list):
        hashtag_list.append(word)


# 가장 많이 나온 단어를 저장
counts = Counter(hashtag_list)
tags = counts.most_common(100) 


# WordCloud 생성
wc = WordCloud(font_path='BMEULJIROTTF.ttf',background_color="white", max_font_size=60)
cloud = wc.generate_from_frequencies(dict(tags))


# 생성된 WordCloud를 test.jpg로 저장
# cloud.to_file('test.jpg')

# 워드 클라우드 출력
plt.figure(figsize=(15, 8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina' #화질 높이기

# open으로 txt파일을 열고 read()를 이용하여 읽는다.
text = open('content.txt').read() 


okt = Okt()

# Okt함수를 통해 읽어들인 내용의 형태소를 분석한다.
sentences_tag = []
sentences_tag = okt.pos(text) 

noun_adj_list = []

unselect_list = ['맛있는','있다','없는','있는','많아서','좋아요','그래','같아요','입니다','같은','좋은','아닌','좋아하는','없다','좋네','좋다','있어','있었던','가능한','같다','이쁜','좋아','작은','아니']


# tag가 명사이거나 형용사인 단어들만 noun_adj_list에 넣어준다.
for word, tag in sentences_tag:

    if tag in [ 'Adjective'] and (len(word) > 1) and (word not in unselect_list):
        noun_adj_list.append(word)


# 가장 많이 나온 단어를 저장한다.
counts = Counter(noun_adj_list)
tags = counts.most_common(500) 


# WordCloud 생성
wc = WordCloud(font_path='BMEULJIROTTF.ttf',background_color="white", max_font_size=60)
cloud = wc.generate_from_frequencies(dict(tags))


# 생성된 WordCloud를 test.jpg로 보낸다.
# cloud.to_file('test.jpg')

# 워드 클라우드 출력
plt.figure(figsize=(15, 8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
