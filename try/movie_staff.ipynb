```python
# staff 추출 함수
import requests
import json

def movie_staff(movie_code):
    
    movie_data = []

    for code in movie_code:

        res = requests.get(STAFF_URL.format(API_KEY, code))
        datas = res.json()

        try:
            movieInfo = datas['movieInfoResult']['movieInfo']

            # 영화코드
            moviecode = code

            for i in movieInfo['staffs']:
                try:
                    # 스텝이름, 역할 추출
                    staffname = i['peopleNm']
                    role = i['staffRoleNm']
                except:
                    # 한국이름없으면 영어이름 가져오기
                    staffname = i['peopleNmEn']
                    role = i['staffRoleNm']

                movie_list=[moviecode,staffname,role]

                movie_data.append(movie_list)

        except:
            # 에러 처리
            pass
        
    return movie_data
```
