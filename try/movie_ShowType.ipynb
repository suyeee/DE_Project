```python
movie_code = list(df['영화코드'].unique())
movie_code
```

```python
import requests

movie_code_3 = ['20204548', '20206061', '20224634', '20211568','20176068']

API_KEY = "126659a3db102258e4a8096d1c949aa8"
STAFF_URL = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key={}&movieCd={}'

datas = []

for code in movie_code_3:
    res = requests.get(STAFF_URL.format(API_KEY, code))
    datas.append(res.json())

# movieInfoResult = datas[0]['movieInfoResult']
# movieInfo = movieInfoResult['movieInfo']
# movieInfo
datas
```

```python
movieInfo = datas[0]['movieInfoResult']['movieInfo']
movieInfo['showTypes']
```

```python
# test- 각 영화별로 영화코드, 상영형태 출력
movie_showtype = []

for k in range(len(datas)):
    movieInfo = datas[k]['movieInfoResult']['movieInfo']
    
    # 영화코드
    moviecode = movieInfo['movieCd']
    
    for i in movieInfo['showTypes']:
        try:
            # 상영형태구분, 상영형태명 추출
            group = i['showTypeGroupNm']
            type = i['showTypeNm']
        except:
            # 상영한적이 없으면 none으로 처리
            group = 'None'
            type = 'None'
            
        movie_list=[moviecode,group,type]
        
        movie_showtype.append(movie_list)
        
movie_showtype
```

```python
pd.DataFrame(movie_showtype, columns=['moviecode', 'showTypeGroup', 'showType']).to_csv('movie_showtype_test1.csv', index = False)
```
